**유용한 Google Colab 단축키**

- 현재 셀 앞에 코드 셀 만들기: Ctrl + M A
- 현재 셀 뒤에 코드 셀 만들기: Ctrl + M B
- 커서를 기준으로 셀 분할: Ctrl + M + -
- 코드 셀 실행: Ctrl + Enter(머물기), Shift + Enter(이동), Alt + Enter(만들기 + 이동)
- 셀 삭제: Ctrl + M D
- 코드 셀을 텍스트 셀로 변환: Ctrl + M Y
- 텍스트 셀을 코드 셀로 변환: Ctrl + M M
