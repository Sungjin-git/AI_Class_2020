## 1.1. 산술 연산자(Arithmetic Operators)

- 대부분의 연산자와 자료형을 우리는 이미 알고 있는 것들입니다.( 2 + 3.5 = 5.5)
- **사칙연산**은 일반적으로 알고 있는 연산 방법(더하기 빼기 등)과 같습니다.
- **나머지**(%), **몫**(//), **제곱**(**)에 대한 부분은 조금 낯설 것이니 관심을 가지고 배워야 합니다.

**1) 산술 연산자 사용**

- 사칙연산자는 이미 알 것이니 넘어가도 될 내용입니다.

```python
# 사칙연산
print(20 + 10)
print(20 - 10)
print(20 * 10)
print(20 / 10)
```

- 조금은 낯선 연산자(나머지, 몫, 제곱)를 확인해 봅시다.

```python
# 나눗셈의 나머지(모듈로)
print(23 % 3)
```

```python
a = 13
if a % 2 == 0:
    print('짝수')
else:
    print('홀수')
```

```python
a = 13
if a % 2 == 0: print('짝수')
else: print('홀수')
```

```python
# 나눗셈의 몫
print(23 / 3)
print(23 // 3)
```

```python
# 나이를 10 단위로 바꾸기/나이 그룹화
a = 23
b = a // 10 * 10
print(b)
```

```python
# 제곱
print(3 ** 2)
print(3 ** 3)
```

**2) print() 함수의 편의성**

- 코드 결과를 효율적으로 확인하기 위해 print() 함수를 많이 사용할 것입니다.
- 아래 두 코드의 실행 결과를 비교해 보면 print() 함수의 필요성을 알 수 있습니다.

## 1.2. 비교 연산자(Comparison Operators)

- = 는 값을 대입하는 연산자이며, '같다' 비교는 <font color='red'>**==**</font> 연산자를 사용합니다.
- '같지않다' 비교는 <> 가 아닌 <font color='red'>**!=**</font> 을 사용해야 함을 기억하세요.
- Python은 문자의 **대문자와 소문자를 구분**하여 비교하니 주의하세요.
- 비교 연산의 결과는 **True** 아니면 **False** 가 됩니다.(true, false가 아닌, True, False 임)

```python
# '같다' 비교
print('abc' == 'abc')
print('abc' == 'Abc')
```

```python
# '같지않다' 비교
print('abc' != 'abc')
print('abc' != 'Abc')
```

```python
# '크다' 비교
print('abc' < 'Abc')
print('abc' < 'bbc')
```

```python
'A' < 'a'
```

```python
print(ord('A'))
print(ord('a'))
print(ord('1'))
```

```python
'1' < 'A'
```

## 1.7. 부울형(Boolean)

- **True**, **False** (첫 문자를 꼭 대소문자로 써야 함) 값을 가지는 자료형입니다.
- True, False와 문자열 'True', 'False'는 전혀 다르니 실수하지 않아야 합니다.
- True와 False는 일반적으로 **조건문**이나 **반복문**에 많이 사용합니다.

**1) True, False**

```python
t = True
f = False
print(type(t))
```

```python
# 조건의 참과 거짓
print(10 > 1)
print(10 == 10)
print(10 > 100)
```

**2) 논리 연산자(Logical Operators): and, or, not**

- 대문자 AND, OR, NOT이 아닌 소문자 and, or, not임에 각별한 주의가 필요합니다.

```python
# 부울형과 논리 연산자
t = True
f = False

print(t and f)
print(t or f)
print(not t)
```

## 1.8. 자료형 사이의 변환

- **float()**, **int()**, **str()**, **bool()** 함수를 사용해 자료형을 변환합니다.
- 함수 이름이 자료형 이름과 같으므로 쉽게 기억할 수 있을 것입니다.
- bool() 함수는 공백 문자('') 와 0에 대해 False를 반환합니다.(있으면 True, 없으면 False로 기억하면 됨) 

**1) float() 함수: 정수 → 실수**

```python
# 정수 --> 실수
a = float(10)
print(type(a), a)
```

**2) int() 함수: 실수 → 정수**

```python
# 실수 --> 정수, 소수점 자리 버림
b = int(123.65)
print(type(b), b)
```

다음 셀에서 변수 b, c의 값을 예측한 후 print() 함수로 확인해 보세요.


**3) str() 함수: 숫자 → 문자열**

```python
# 숫자 --> 문자열
c = str(123.45)
print(type(c), c)
```

**4) bool() 함수: 숫자 → 부울형**

- 0이면 False, 나머지는 True

```python
# 숫자 --> 부울형
d = bool(-12)
print(type(d), d)

e = bool(0)
print(type(e), e)
```

**5) bool() 함수: 문자 → 부울형**

- 스페이스 문자도 없는 완전 공백('')이면 False, 나머지는 True

```python
# 문자 --> 부울형
f = bool('a')
print(type(f), f)

g = bool('')
print(type(g), g)

h = bool(' ') # ''와 ' '는 다름
print(type(h), h)
```

```python
# 값의 유무 판단
a = ''
if a:
    print(a)
else:
    print('hul')
```

## 1.9. 날짜 시간 자료형

- Python은 날짜와 시간 관련 자료형을 제공하지 않습니다.
- 필요하다면 **datetime** 라이브러리를 불러와 사용해야 합니다.

```python
# 라이브러리 불러오기
from datetime import datetime

# 현재 날짜와 시간 얻기
dt = datetime.now()
```

```python
# 날짜 시간 정보 표시
print('날짜와 시간: ' + str(dt))
print('년: ' + str(dt.year))
print('월: ' + str(dt.month))
print('일: ' + str(dt.day))
print('시: ' + str(dt.hour))
print('분: ' + str(dt.minute))
print('초: ' + str(dt.second))
print('날짜: {}-{}-{}'.format(dt.year, dt.month, dt.day))
```
