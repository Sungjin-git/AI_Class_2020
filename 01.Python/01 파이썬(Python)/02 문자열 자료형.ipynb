## 2.1. 문자열 사용

- Python은 문자열에 대해 상당한 편의성을 지원합니다.

**1) 문자열 표현**

- **작은 따옴표**, 또는 **큰 따옴표**를 사용해 문자열 표시합니다.
- **len()** 함수로 문자열 길이(=포함된 문자 개수)를 확인할 수 있습니다.

```python
# 기본적인 문자열 표현 방식
hello = 'hello'
world = "world"
print(hello)
print(world)
```

```python
# 문자열 길이 확인
len(hello)
```

**[참고] 나열된 문자열 사이에 콤마 있고 없음의  큰 차이점**

```python
# 문자열을 콤마 없이 나란히 입력하면 합쳐짐
hello = 'hello' 'world'
hello
```

```python
# 문자열을 콤마로 구분하여 나열하면 튜플(Tuple)이 됨
# [참고] 튜플: (1, 2 ,3) 형태로 괄호 안에 요소가 나열된 자료형
hello = 'hello', 'world'
hello
```

**2) 여러 줄 문자열 표현**

- 문자열 앞뒤에 따옴표 세 개를 나열하거나, \\n을 사용해 여러 줄 표현이 가능합니다.
- print() 함수를 사용해야 의도한 내용이 제대로 표시됩니다.

```python
# 여러 줄 문자열 표시
s1 = "여러 줄의 문자열을 표시할 때는 \n을 사용하세요"
print(s1)
```

```python
s2 = """또 다른 방법은 
큰 따옴표나 작은 따옴표를 세 개 사용하는 것입니다."""
print(s2)
```

**3) 문자열에 따옴표 넣기**

- 큰 따옴표 안에 작은 따옴표, 또는 작은 따옴표 안에 큰 따옴표를 사용합니다.
- \\' 또는 \\\"를 사용해도 됩니다.
- print() 함수를 사용해야 의도한 내용이 제대로 표시됩니다.

```python
# 문자열에 따옴표 넣기
s3 = "문자열 안에 따옴표, 예를 들어 '를 넣고 싶을 때는 이렇게 하세요."
print(s3) 
```

```python
s4 = "아니면 \' 또는 \"를 사용해도 됩니다."
print(s4)
```

```python
# 문자열에 \ 넣기
s5 = "아니면 \\' 또는 \\\"를 사용해도 됩니다."
print(s5)
```

## 2.2. 문자열 연산

- 문자열 연산은 숫자 연산과 당연히 다릅니다.
- 문자열 연산에 대해 알아두며 이후에 편리하게 사용할 수 있습니다.

**1) 문자열 + 문자열 = 문자열 결합**

```python
# + 사용해 문자열 결합
hello = 'hello' 
world = 'world'
hw = hello + ' ' + world
hw
```

**2) 문자열 * 정수 = 문자열 반복**

```python
# * 사용해 문자열 반복
print('*' * 17)
print('** ' + hello + ' ' + world + ' **')
print('*' * 17)
```

```python
# 문자열에 0을 곱하면 문자열이 제거됨
hello = 'hello'
print(hello * 1)
print(hello * 0)
print(hello * 2)
```

## 2.3. format 메소드 사용

- 값을 원하는 위치에 원하는 형태로 표시하고자 할 때 print() 함수 안에서 많이 사용합니다.

### 2.3.1. 방식의 변화


- 시대가 변하면 말이 변하듯, 기능 향상과 편의성을 위해 코딩 규칙이 바뀌는 경향이 있습니다.
- 가능하면 최근에 사용되는 방식을 사용하기를 권고합니다.

**1) 이전 방식**

- 너무 기호화된 느낌이 듭니다.

```python
'%s %s - %d' % ('hello' , 'world', 12) 
```

**2) 요즘 방식**

- 좀더 간결해진 느낌입니다.

```python
'{} {} - {}'.format('hello' , 'world', 12)
```

- 변수를 사용해도 됩니다.

```python
fmt = '{} {} - {}'
print(fmt.format('hello' , 'world', 12))
```

```python
print('My age = {}, Your age = {}'.format(32, 31))
```

**3) 표시 위치 지정**

- { } 안에 변수 위치 값을 지정해야 하나 순서대로 사용될 것이면 생략하는 경우가 많습니다.
- 순서를 바꾸어 사용할 때는 적절한 위치 값을 { } 안에 지정합니다.

```python
hello = 'hello' 
world = 'world'

# {} 안에 위치 값 지정 --> 지정한 위치의 변수가 대입 됨
print('{0} {1} {2}'.format(hello, world, 12)) 
print('{2} {0} {1}'.format(hello, world, 12))
```

```python
# 문자열에 {} 포함시켜 표시하기
print('{} {} {{}}'.format(10, 20))
```

## 2.4. 문자열 인덱싱과 슬라이싱

- **인덱싱(Indexing)**: 정수형 위치 인덱스를 사용해 원하는 요소를 조회합니다.
- **슬라이싱(Slicing)**: 범위를 지정해 일정 범위의 요소를 조회합니다.
- <font color="red">인덱스 값은 0부터 시작</font>함을 잊어서는 안됩니다.


- 문자열은 문자 요소가 모여서 만들어진 자료형, 즉 **'ABC'는 'A', 'B', 'C' 세 요소가 모인 것**입니다.
- 문자열 인덱싱과 슬라이싱은 이후에 다룰 리스트(List)와 비슷한 특성을 가집니다.

**1) 문자열 인덱싱**

<img src='https://raw.githubusercontent.com/jangrae/img/master/str_indexing.png' width=560 align="left"/>

```python
a = 'ABCDEFGHIJKLMN'
#    01234567890123
a[-1]
```

```python
a[len(a) - 1]
```

```python
a_list = list(a)
a_list[3] = 'X'
print(''.join(a_list))
```

```python
# 문자열 인덱싱
s = "나의살던고향은꽃피는산골"

print(s[0])
print(s[3])
print(s[11])
print(s[-1])
print(s[-5])
```

```python
s = "솔미파솔라라솔솔도미레도레"
```

**2) 문자열 슬라이싱**

<img src='https://raw.githubusercontent.com/jangrae/img/master/str_slicing.png' width=450 align="left"/>

```python
# 문자열 슬라이싱
s = "나의살던고향은꽃피는산골"  

print(s[4:6])
print(s[7:])
print(s[-5:])
print(s[-5:len(s)])
print(s[:])
```

## 2.5. 문자열 관련 주요 메소드
* object(변수) 뒤에 '.함수' 형태로 사용하는 함수를 메소드(Method)라고 부릅니다.
- 아래 언급된 메소드 외에도 문자열 관련 메소드가 상당히 많으니 관련 자료를 꼭 참고하기를 권고합니다.

- **upper()** 메소드: 모든 문자를 대문자로 변환합니다.
- **lower()** 메소드: 모든 문자를 소문자로 변환합니다.

```python
s = "I have a Dream."

# 함수 사용
print(s.upper())
print(s.lower())
```

- **replace(x, y)** 메소드: 문자열에 포함된 일부 문자열 x를 문자열 y로 변경합니다.

```python
# 문자열 변경
s = 'nowhere'
s.replace('now', 'now ')
```

```python
# ★ 아래도 기억하자
s = s.replace('now', 'now ')
s
```

- **split((x)** 메소드: x 문자를 구분자로 하여 문자열을 분리하여 리스트 반환합니다.(공백이 기본 구분자)

```python
# 문자열 분리 
s = 'I have a Dream!'
s.split()
```

```python
# 문자열 분리 
s = '2020-03-20'
s.split('-')
```

```python
# 응용
s = '2020-03-20'
year = s.split('-')[0]
month = s.split('-')[1]
day = s.split('-')[2]
print(year, '년', month, '월', day, '일')
```

- **strip(x)** 메소드: x 문자 주변의 공백을 제거합니다.

```python
s = '12월 25일'
# 슬라이싱 안하고 25 뽑기
s.split()[1].split('일')[0]
```

```python
# split과 슬라이싱 사용
s.split()[0][0:2]
```

```python
# 문자열 분리 
s = '    I have a Dream! '
s = s.strip()
print(s)
```

```python
s = '@@@@@I have a Dream!@@'
s = s.lstrip('@')  # 왼쪽 제거 lstrip, 오른쪽 rstrip
s
```
