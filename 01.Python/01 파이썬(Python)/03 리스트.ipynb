## 3.1. 리스트 만들기

### 3.1.1. 다양한 형태의 리스트

**1) 일반 변수와 리스트 비교**

- 일반 변수: 50명 학생 성적을 관리하려면 50개 변수가 필요합니다.
- 리스트: 50명 학생 성적을 리스트 하나로 관리할 수 있습니다.

```python
# 개별 값을 갖는 기존 변수
score1 = 85
score2 = 90
score3 = 80
score4 = 75
score5 = 95
```

```python
# 변수들의 합 구하기
score_sum1 = score1 + score2 + score3 + score4 + score5
score_sum1
```

```python
# 리스트 만들기
score = [85, 90, 80, 75, 95]
score
```

```python
# 리스트 요소의 합
score_sum2 = sum(score)
score_sum2
```

```python
# 리스트 요소 중 가장 큰 값
score_max1 = max(score)
score_max1
```

```python
# score1 ~ score5 중 가장 큰 값을 구할 수 있는 분?
max_num = score[0]
for num in score:
    if max_num < num:
        max_num = num
print(max_num)
```

```python
a = [1, 2, 3, 4, 5, 6]
a
```

```python
# 리스트를 배우는 목적 1 : 묶어서 처리하기
sum(a)
```

```python
# 리스트를 배우는 목적 2 :순차적인 처리
for num in a:
    if num % 2 == 0:
        print(num)
```

### 3.1.2. range() 함수로 리스트 만들기

- **range(시작값, 종료값, 증가값)** 형태로 지정합니다.
- 시작값, 종료값, 증가값은 모두 **정수**만 사용이 가능합니다.
- 함수 결과에 종료값은 포함되지 않고 **종료값 - 1 까지의 숫자**가 만들어집니다.

**1) 기본 사용법**

- **range(*n*)**: 0 부터 n-1 까지의 정수 범위가 됩니다.
- **range(*m*, *n*)**: m 부터 n-1 까지의 정수 범위가 됩니다.

```python
# 일정 범위의 값을 갖는 리스트 만들기
nums = list(range(10))
nums
```

```python
# 위 결과와 동일
nums = list(range(0, 10))
nums
```

```python
# 음수 사용 가능
nums = list(range(-5, 6))
nums
```

**2) 증가값 지정**

- **range(*m, n, x*)**: m 부터 n-1 까지 x 만큼씩 증가하는 정수가 됩니다.

```python
# 2씩 증가
nums = list(range(0, 10, 2))
nums
```

```python
# 3씩 증가
nums = list(range(0, 10, 3))
nums
```

```python
# 1씩 감소
nums = list(range(9, -1, -1))
nums
```

```python
# 1 ~ 100 정수 리스트 a 만들기
a = list(range(1, 101))
a
```

```python
s = 0
for num in range(1, 11):
    if num % 2 == 0:
        s += num
print(s)
```

- **0부터 시작**하는 **정수형 위치 인덱스**를 사용해 원하는 위치의 요소를 확인합니다.
- 0부터 시작함이 매우 낯설지만 자주 사용해 익숙해져야 합니다.

## 3.3. 리스트 슬라이싱(Slicing)
- 대괄호안에 [m:n] 형태로 지정하여 범위를 지정하여 요소를 확인합니다.
- m 부터 n-1 까지, 즉 **n 번째 요소는 결과에 포함되지 않습니다.**

<img src='https://raw.githubusercontent.com/jangrae/img/master/list_indexing.png' width=540 align = "left"/>

```python
# 리스트 만들기
member = ['홍길동', 85, '한사랑', 95, '일지매', 90,' 박여인', 75]

print(member[2:6])
print(member[6:])
print(member[:4])
print(member[:])
print(member[0:-2])
```

```python
# 리스트를 요소로 갖는 리스트 예
score = [80, [90, 100, 85], 85, 90]
score[1][1:3]
```

## 3.4. 리스트 연산

- 리스트 연산은 문자열 연산과 같은 결과를 보여줍니다.
- 리스트 + 리스트 = 리스트 연결
- 리스트 * 정수 = 리스트 반복

**1) 리스트 더하기**

```python
# 두 리스트 더하기
score1 = [90, 80, 95, 70]
score2 = [70, 75, 90, 100]
score3 = score1 + score2
score3
```

```python
# 리스트에서 리스트 빼기?
score1 = [90, 80, 95, 70]
score2 = [70, 75, 90, 100]
score3 = score1 - score2
score3
```

**2) 리스트 반복**

```python
# 리스트 요소를 곱한 수 만큼 반복하기
score1 = [90, 80, 95, 70]
score2 = score1 * 2
score3 = score1 * 3
print(score1)
print(score2)
print(score3)
```

## 3.6. 리스트 복사

- 리스트 = 리스트 형태는 우리가 생각하는 복사가 아닙니다.
- **범위를 지정한 복사가 진정한(?) 복사입니다.**

```python
# 리스트 복사
nums = list(range(10))
nums_copy_1 = nums                  # 그냥 복사
nums_copy_2 = nums[:]               # 전체 범위를 지정하여 복사
print(nums_copy_1)
print(nums_copy_2)
```

```python
# copy_1의 첫번째 값 변경 후 원본과 비교
nums_copy_1[0] = 'a'
print(nums_copy_1)
print(nums_copy_2)
print(nums)
```

```python
# copy_2의 첫번째 값 변경 후 원본과 비교
nums_copy_2[0] = 'b'
print(nums_copy_1)
print(nums_copy_2)
print(nums)
```
