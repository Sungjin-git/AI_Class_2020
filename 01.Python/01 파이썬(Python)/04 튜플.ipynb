# 4. 튜플(Tuple) 자료형

- 튜플은 대괄호([ ])가 아닌 **소괄호(( ))**를 사용해서 표현합니다.
- 튜플은 한 번 만들면 **요소의 값을 바꿀 수 없습니다.**(리스트와 다른 점)
- 그 외는 리스트와 같은 특성을 가지므로 이해하기 쉬울 것입니다.
- 요소를 변경할 수 없어 지원되는 메소드도 거의 없습니다.


## 4.1. 튜플 만들기

- 리스트는 대괄호[[ ]]이고, 튜플은 소괄호(( ))인 사실만 기억하면 됩니다.

**1) 빈 튜플, 요소가 하나인 튜플 만들기**

- 요소가 하나인 튜플을 만들 때는 주의해야 합니다.

```python
# 빈 튜플 만들기
score = ()
score
```

```python
# 튜플을 잘못 만든 예, 주의!!!, 아래 예시에서 ()는 있으나 마나
score = (90)
print(score)
print(type(score))
```

```python
# 요소가 하나인 튜플 만들기
score = (90,)
print(score)
print(type(score))
print(len(score))
```

**2) 다양한 형태의 튜플**

- 모든 방법이 리스트와 같습니다.
- 단, 괄호를 생략할 수 있다는 점은 리스트와 다릅니다.(리스트에서는 대괗호 생략 불가)

```python
# 여러 형태의 튜플 만들기
score1 = (90, 85, 70)
score2 = 90, 85, 70   #()를 생략할 수 있다
score3 = (90, 85, 70, ('A', 'B', 'D', 'F'))
print(score1)
print(score2)
print(score3)
```

```python
# range() 함수를 사용해 만들기
nums = tuple(range(5))
print(nums)
```

```python
# 문자열로 튜플 만들기
chars = tuple('PYTHON')
print(chars)
```

**[참고] 튜플 활용**

- **괄호를 생략**할 수 있는 튜플의 특징을 이용하면 편리한 부분이 있습니다.

```python
# 괄호 없이 튜플 만들기
a = 10
b = 20
c = a, b
c
```

- 여러 변수에 동시에 값을 대입할 수 있습니다.

```python
# 여러 변수에 동시에 값 대입
a, b, c = 10, 20, 30
print('a =', a)
print('b =', b)
print('c =', c)
```

```python
# 여러 변수에 동시에 값 대입(튜플 포함)
a, b, c = 10, 20, (1, 2, 3)
print('a =', a)
print('b =', b)
print('c =', c)
```

- 두 변수값을 교환할 수 있습니다.

```python
# 변수 값 서로 변경
a, b = 10, 20
print('a =', a, 'b =', b)
a, b = b, a
print('a =', a, 'b =', b)
```

- 함수 리턴값을 두 변수에 대입할 수 있습니다.

```python
# 기본 연산
a, b = 45, 20
x = a // b
y = a % b
print(x, y)
```

```python
# 튜플 사용
a, b = 45, 20
x, y = divmod(a, b)
print(x, y)
```

## 4.2. 튜플 인덱싱과 슬라이싱

- 앞에서 다룬 리스트와 방법과 결과가 같으니 쭈~욱 보고 넘기면 됩니다.

**1) 튜플 인덱싱**

```python
# 2번째 요소 확인
score = (90, 85, 70, 95)
print(score[1])
```

```python
# 4번째 요소인 튜플 확인
score = (90, 85, 70, ('A', 'B', 'D', 'F'))
score[3]
```

```python
# 4번째 요소 튜플의 요소 확인
score = (90, 85, 70, ('A', 'B', 'D', 'F'))
print(score[3][1])
print(score[3][2])
print(score[3][3])
```

**2) 튜플 슬라이싱**

```python
# 2번째 요소부터 끝까지
score = (90, 85, 70, 95)
score[1:]
```

## 4.3. 튜플 연산

- 앞에서 다룬 리스트와 연산 방법과 결과가 같으니 역시 쭈~욱 넘기면 됩니다.

```python
# 두개의 튜플 더하기
score1 = (90, 85, 70, 95)
score2 = (80, 85, 65, 90)
score3 = score1 + score2
score3
```

```python
# 튜픙 요소 여러번 반복
score1 = (90, 85, 70, 95)
score2 = score1 * 3
score2
```
