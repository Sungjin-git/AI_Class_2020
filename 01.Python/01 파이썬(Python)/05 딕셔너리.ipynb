# 5. 딕셔너리(Dictionary) 자료형

- {Key1: Value1, Key2: Value2, Key3, Value3}의 형태를 가집니다.
- 딕셔너리는 순서가 의미가 없으며, **키(Key)를 사용해 값(Value)을 확인합니다.**
- 리스트와 더불어 많이 사용되는 자료형이니 필히 익숙해져야 합니다.

## 5.1. 딕셔너리 만들기

### 5.1.2. 딕셔너리로 자료형 변환

- **dict() 함수**를 사용해 Key:Value 형태를 가질 수 있는 자료형을 딕셔너리로 변환합니다.
- 대부분 리스트로 딕셔너리를 만듭니다.

**1) 리스트 → 딕셔너리 변환**

```python
# 리스트 만들기
members = [['홍길동', 100], ['일지매', 90], ['한여인', 90], ['강우동', 95]]

# 딕셔너리로 변환
member_dic = dict(members)
member_dic
```

**2) 튜플 → 딕셔너리 변환**

```python
# 튜플 만들기
members = (('홍길동', 100), ('일지매', 90), ('한여인', 90), ('강우동', 95))

# 딕셔너리로 변환
member_dic = dict(members) 
member_dic
```

## 5.2. 딕셔너리 조회

- 딕셔너리는 요소 순서가 없으므로 **정수형 위치 인덱스를 사용할 수 없습니다.**
- **Key를 사용해 Value를 찾는 방식으로 조회합니다.**
- in을 사용해 해당 Key가 있는지 확인 가능합니다.

```python
# 딕셔너리 만들기
member = {'이름': '홍길동', 
          '취미': ['독서', '여행', '걷기']}
member
```

```python
# 이름 확인
member['이름']
```

```python
# 취미 확인
member['취미']
```

```python
# 같은 key가 있을 경우 
member = {'이름': '홍길동', 
          '이름': '일지매', 
          '취미': ['독서', '여행', '걷기']}
```

```python
member['이름'] 
```

```python
# value 가 있는지 확인
'이름' in member    # True
```

```python
'성격' in member    # False
```

```python
# in 연산자는 딕션너리에 국한되지 않음
a = [1, 2, 3, 4, 5, 6]
5 in a
```

## 5.3. 딕셔너리 변경, 추가와 삭제

- 인덱스 대신 Key로 찾아 변경한다는 것 외에는 리스트와 같습니다.

**1) 변경, 추가**

```python
# 딕셔너리 만들기
member = {'이름': '홍길동', 
          '나이': 20}
member
```

```python
# 딕셔너리 요소 값 변경: 찾기 + 바꾸기
member['나이'] = 30
member
```


- 없는 Key로 변경하면 Key:Value 형태로 추가됩니다.

```python
# 딕셔너리에 요소 추가: 찾기 + 추가
member['취미'] = ['독서', '여행']
member
```

```python
# 딕셔너리에 요소 추가 - value가 리스트인 경우 특정값 바꾸기
member['취미'][1] = '해외여행'
member
```

```python
# 취미 추가
member['취미'].insert(1, '요리')
member
```

**2) 삭제**

- **del** 문으로 Key를 지정해 해당 요소를 삭제합니다.
- 모든 요소 삭제는 **딕셔너리 = { }** 형태로 수행합니다.

```python
# 딕셔너리 만들기
member = {'이름': '홍길동', 
          '나이': 20, 
          '지역':'서울',
          '성별': 'M',
          '등급': 'Gold'}
member
```

```python
# 딕셔너리에서 지정한 요소 삭제
del member['나이']
member
```

```python
# pop() 메소드로 딕셔너리에서 지정한 요소 삭제
del_member = member.pop('지역')
print(del_member)
print(member)
```

```python
# 딕셔너리에서 임의 요소 삭제
del_member = member.popitem()
print(del_member)
print(member)
```

```python
# 딕셔너리에서 모든 요소 삭제
member= {}
member
```

```python
# 집합
a = {1, 2, 3, 4}
b = {1, 2, 5, 6}
type(a)    # set
```

```python
# 집합을 {}로 초기화 하면 딕셔너리가 됨
a = {}
type(a)    # dict
```

## 5.4. 딕셔너리 관련 메소드

**1) Key 확인**

- **keys()** 메소드를 사용합니다.
- 모든 key를 요소로 갖는 dict_keys 개체를 반환합니다.

```python
# 딕셔너리 만들기
member = {'이름': '홍길동', 
          '취미': ['독서', '여행', '걷기']}
member
```

```python
# key 확인
member.keys()
```

**2) Value 확인**

- **values()** 메소드를 사용합니다.
- 모든 value를 요소로 갖는 dict_values 개체를 반환합니다.

```python
# value 확인
member.values()
```

**3) Key, Value 확인**
- **items()** 메소드를 사용합니다.
- 모든 key와 value를 (key, value) 형태로 갖는 dict_items 개체를 반환합니다.

```python
# 튜플 형태로 표시
member.items()
```

```python
# items의 결과는 튜플로 받을 수 있다
for a, b in member.items():
    print(a, '-', b)
```

```python
# items가 없더라도 같은 결과를 만들 수 있다
for a in member.keys():
    print(a, '-', member[a])
```

**4) 모든 요소 삭제**
- **clear()** 메소드를 사용합니다.
- 모든 요소를 삭제합니다.

```python
# 모든 요소 삭제
member.clear()
member
```

**5) 찾기(없을 때 대신할 값 지정)**

- **get()** 메소드를 사용합니다.
- key로 value를 찾을 때 없는 경우 대신 표시할 값을 지정합니다.

```python
# 딕셔너리 만들기
member = {'이름': '홍길동', '취미': ['독서', '여행', '걷기']}
member
```

```python
try:
    member['성격']
except KeyError as e:
    print('아 짜증 에러')
```

```python
# 이름 확인
member.get('이름')
```

```python
# 이메일 확인
member.get('이메일', '없음')
```

```python
member.get('성격', '특이점 없음')
```

- 대응되는 value를 찾지 못할 경우 반환되는 값을 지정할 수 있습니다.

```python
# 이메일 확인, 없으면 '없음' 표시
member.get('이메일', '없음')
```

## [실습] 딕셔너리

```python
stu01 = {
    'no': 1,
    'name': '홍길동',
    'score': [92, 96, 98]
}
stu02 = {
    'no': 2,
    'name': '한사랑',
    'score': [98, 64, 100]
}
stu03 = {
    'no': 3,
    'name': '일지매',
    'score': [88, 84, 92]
}
print(stu01)
print(stu02)
print(stu03)
```

1) 각 학생들의 번호에 10씩 더하세요.

```python
# 10 더하기
stu01['no'] = stu01['no'] + 10
stu02['no'] = stu02['no'] + 10
stu03['no'] = stu03['no'] + 10

# 확인
print(stu01)
print(stu02)
print(stu03)
```

2) 각 학생별로 점수 평균을 구하세요.

```python
# 평균 구하기
stu01_mean = sum(stu01['score']) / len(stu01['score'])
stu02_mean = sum(stu02['score']) / len(stu02['score'])
stu03_mean = sum(stu03['score']) / len(stu03['score'])

# 확인
print(stu01_mean)
print(stu02_mean)
print(stu03_mean)
```

3) 아래 out을 이용하여 stu01, stu02, stu03의 번호, 이름, 평균점수를 출력하세요.

```python
# 출력
out = "번호: {}, 이름: {}, 평균점수: {}"

print(out.format(stu01['no'], stu01['name'], stu01_mean))
print(out.format(stu02['no'], stu02['name'], stu02_mean))
print(out.format(stu03['no'], stu03['name'], stu03_mean))
```

4) 평균점수는 소수점 이하 2자리만 보이게 수정해서 출력하세요.

```python
# 소수점 두 자리만 보이게
out = "번호: {}, 이름: {}, 평균점수: {}"
print(out.format(stu01['no'], stu01['name'], round(stu01_mean, 2)))
print(out.format(stu02['no'], stu02['name'], round(stu02_mean, 2)))
print(out.format(stu03['no'], stu03['name'], round(stu03_mean, 2)))
```
