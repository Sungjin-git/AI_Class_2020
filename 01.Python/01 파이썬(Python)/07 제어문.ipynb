# 7. 제어문(Control Statements)

- **조건 처리**나 **반복 처리**를 통해 코드의 실행 흐름을 제어할 수 있습니다.
- 다른 언어와 마찬가지로 if문, for문, while문이 존재합니다.
- Python 제어문은 **들여쓰기로 블럭을 지정**하니 이에 주의해야 합니다.
- 다양한 로직 작성을 위해 필히 익혀야 하는 내용이며, 확장 문법을 알아두면 편리합니다.

## 7.1. 조건문: if 문

**1) 단순 if 문**

- 조건이 참인 경우만 처리합니다.

<img src='https://raw.githubusercontent.com/jangrae/img/master/if.png' width=200 align="left"/>

```python
my_score = 73

# 80보다 크거나 같으면 10 더하기 
if my_score >= 80:
    my_score += 10

# 결과 출력
my_score
```

```python
score_list = [75, 80, 70, 95, 90]

# 리스트에 요소가 있으면 출력, 없으면 말고 
if score_list:
    print(score_list)
```

**2) if ~ else 문**

- 조건이 참인 경우와 참이 아닌 경우 각각 처리합니다.

<img src='https://raw.githubusercontent.com/jangrae/img/master/if_else.png' width=200 align="left"/>

```python
my_score = 73

# 80보다 크거나 같으면 Pass, 아니면 Fail 
if my_score >= 80:
    my_result = 'Pass'
else:
    my_result = 'Fail'

# 결과 출력    
my_result
```

```python
score_list = [75, 80, 70, 95, 90]

# 리스트에 요소가 있으면 출력, 없으면 '요소 없음' 출력 
if score_list:
    print(score_list)
else:
    print('요소 없음')
```

- 처리할 구문이 한 줄이라면 다음과 같이 작성할 수 있습니다.

```python
score_list = [75, 80, 70, 95, 90]

# 한 줄로 작성하기
if score_list: print(score_list)
else: print('요소 없음')
```

- 처리할 구문이 없거나 이후에 작성할 목적으로 **pass**를 사용할 수 있습니다.

```python
score_list = [75, 80, 70, 95, 90]

# 우선 pass 
if score_list:
    pass
else:
    print('요소 없음')
```

**3) if ~ elif ~ else 문**

- 여러 조건 중에서 참인 경우 각각 처리, 그렇지 않은 경우 처리합니다.

<img src='https://raw.githubusercontent.com/jangrae/img/master/elif.png' width=200 align="left"/>

```python
my_score = 76

# 점수에 따른 학점 주기 
if my_score >= 90:
    my_grade = 'A'
elif my_score >= 80:
    my_grade = 'B'
elif my_score >= 70:
    my_grade = 'C'
elif my_score >= 60:
    my_grade = 'E'
else:
    my_grade = 'F'

# 결과 출력
my_grade
```

2) score1, score2, score3 점수에 대하여 다음 조건이 처리되도록 조건문을 완성하세요. 
- 하나라도 60점 이하거나 점수의 평균이 70점 이하이면 Fail
- 그렇지 않은 경우에는 Pass

```python
score1 = 60
score2 = 100
score3 = 76

# 조건문 작성
if (score1 <= 60) or (score2 <= 60) or (score3 <= 60):
    print('Fail')
elif ((score1 + score2 + score3) / 3) <= 70:
    print('Fail')
else:
    print('Pass')
```

## 7.3. 반복문: while 문

- 조건문이 True인 동안 while문 안의 문장이 반복해서 실행됩니다.
- 특정 상황에서 반복을 중지하게 while 문 안에서 조건을 제어합니다.

**1) 일반 while 문**

- 문자열 길이를 반환하는 len() 함수가 리스트의 요소 개수를 반환합니다.

```python
fruits = ['apple', 'banana', 'cherry']

i = 0 
while i < len(fruits):
    print(fruits[i])
    i += 1
```

**2) 무한루프 다루기**

- break 문: 반복문 안에서 특정 조건이 되어 반복문을 빠져 나올 때 사용합니다.
- continue 문: 반복문을 중단시키지 않고 다음 반복으로 넘어갈 때 사용합니다.

```python
# 100까지 짝수 합 구하기
tot, i = 0, 0

while True:
    i += 1
    if i > 100:
        break
        
    if i % 2 == 0:
        tot += i
    else:
        continue
   
tot
```

- input() 함수를 사용해 사용자의 입력을 받을 수 있습니다.
- isnumeric() 메소드: 숫자이면 True 아니면 False 반환합니다.

```python
# 문자를 입력하면 continue, 0을 입력하면 break, 아니면 10으로 나눈 나머지 표시
while True:
    response = input('숫자를 입력하세요(0 = 종료):')
    if not response.isnumeric():
        continue
    if int(response) == 0:
        break
    result = int(response) % 10
    print("{} 나누기 10의 나머지는 {}입니다.".format(response, result))
```

## 7.4. 리스트와 for 문

- 반복문과 가장 많이 사용되는 자료형이 리스트형입니다.
- 반복문 작성이 익숙해지도록 학습과 실습이 필요합니다.

**1) 기본 반복문**

- 첫 번째 요소부터 마지막 요소까지를 대상으로 코드를 반복 처리합니다.

```python
# 반복문(Loops)
member = ['홍길동', '한사랑', '일지매', '박여인', '강우동']

# 루프 돌면서 이름 출력
for mem in member:
    print(mem)
```

**2) enumerate() 함수 사용**

- 영어 단어 enumerate의 뜻은 '**낱낱이 세다**' 입니다.
- 순서가 있는 자료형를 입력받아, 인덱스와 요소를 포함하는 오브젝트로 반환합니다.
- 인덱스와 요소를 반환하여 몇 번째, 어느 요소를 처리하고 있는지 확인 가능하게 합니다.

```python
# '인덱스: 이름' 형태로 출력
for idx, mem in enumerate(member):
    print('{}: {}'.format(idx, mem))
```

```python
# 인덱스 값이 짝수인 경우만 '인덱스: 이름' 형태로 출력
for idx, mem in enumerate(member):
    if idx % 2 == 0:
        print('{}: {}'.format(idx, mem))
```

```python
a = tuple(enumerate(member))
a
```

```python
# 문자열과 enumerate
for idx, val in enumerate('나의살던고향'):
    print(idx, val)
```

**3) 반복문 확장**

- 이 방법을 사용하면 반복문 코드가 매우 간결해집니다.
- 처음에는 다소 낯설 수 있으나, 이것이 **Python 스러운 문법**이라 할 수 있습니다.

```python
# 반복문 조금 더 살펴보기
nums = [0, 1, 2, 3, 4]

# 빈 리스트 만들기
squares = []

# 2를 곱한 결과를 squares 리스트에 추가
for x in nums:
    squares.append(x ** 2)

# 확인
squares
```

```python
# 반복문 확장 문법 #1
squares = [x ** 2 for x in nums]
squares
```

```python
# 반복문 확장 문법 #2 
# 반복문 + 조건문: 짝수인 값의 제곱을 구해서 even_squares 리스트에 넣기
even_squares = [x ** 2 for x in nums if x % 2 == 0] 
even_squares
```

```python
a = [x + 10 for x in range(11) if x % 2 == 0]
a
```

```python
# 0 ~ 10 중에서 5보다 큰 값만 추출
a = [x for x in range(11) if x > 5]
a
```

```python
a = (x for x in range(11) if x > 5)
a
```

## 7.5. 딕셔너리와 for 문

**1) 기본 반복문**

```python
# 딕셔너리 만들기
score = {'홍길동': 85, '한사랑': 95, '일지매': 90}

# key와 value 조회
for key in score:
    value = score[key]
    print('{}: {}'.format(key, value))
```

**2) items() 메소드 사용**

- 딕셔너리를 반복문에 사용할 때 key와 value를 반환해주는 items() 메소드를 많이 사용합니다.

```python
# items() 메소드 사용
for key, value in score.items():
    print('{}: {}'.format(key, value))
```

학생 이름과 점수를 갖는 아래 딕셔너리에 대해 반복문을 작성해 다음과 같이 학생들 이름과 점수 합이 출력되게 해보세요.

~~~
홍길동: 350
한사랑: 355
일지매: 300
~~~

```python
# 딕셔너리 만들기
score = {'홍길동': [85, 80, 90, 95], 
         '한사랑': [95, 90, 80, 90], 
         '일지매': [90, 70, 60, 80]}
         
# 이름과 점수 출력
for name, point in score.items():
    print(name, point)
```

**3) 반복문 확장**

- 앞에서 '리스트와 for 문'에서 다룬 내용과 동일하게 딕셔너리에 대해 사용할 수 있습니다.

```python
# 딕셔너리 만들기
score = {'홍길동': 85, '한사랑': 95, '일지매': 90}
score
```

```python
# 확장 문법 - 반복문 + 제어문
score_over_90 = {k: v for k, v in score.items() if v > 90}
score_over_90
```

- 다음과 같은 코드로 같은 결과를 얻을 수 있습니다.

```python
# items() 메소드 사용 안하면
score_over_90 = {k: score[k] for k in score if score[k] > 90}
print(score_over_90) 
```
