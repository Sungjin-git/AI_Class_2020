# 8. 함수(Function)

## 8.1. 함수 구조

```python
# 두 매개변수의 합을 반환하는 함수 만들기
def hap(x, y):
    return x + y
```

```python
# 함수 사용
a = 10
b = 20
c = hap(a, b)
print("{0}와 {1}의 합은 {2}".format(a, b, c))
print("{0}와 {1}의 합은 {2}".format(10, 20, hap(10, 20)))
```

```python
# 매개변수가 Negative 인지, Positive 인지, 0 인지 확인하는 함수
def sign(x):
    if x > 0:
        result = 'Positive'
    elif x < 0:
        result = 'Negative'
    else:
        result = 'Zero'
    return result

# 함수 사용
for x in [-15, 0, 12]:
    print('{:3d}: {}'.format(x, sign(x)))
```

두 문자열을 매개변수로 받아 두 문자열을 연결해 반환하는 concat 함수를 만들어 사용해 보세요.

```python
# 함수 만들기
def concat(string1, string2):
    return string1 + string2

# 함수 사용
concat('ABC', 'DEF')
```

```python
# 함수 만들기 : 나 혼자 해보기
def concat(*string):
    y = ''
    for x in string:
        y = y + x
    return y

# 함수 사용
concat('ABC', 'DEF', '123')
```

```python
# 함수 만들기 : 나 혼자 해보기
def concat(*string):
    return ''.join(string)

# 함수 사용
concat('ABC', 'DEF', '123', '+_)')
```

리스트를 매개변수로 받아 짝수만 반환하는 even_num 함수를 만들어 사용해 보세요.

```python
# 함수 만들기
def even_num(num_list):
    result = [x for x in num_list if x % 2 == 0]
    return result

# 함수 사용
a = [1, 2, 3, 4, 5, 6]
a = even_num(a)
a
```

```python
# 함수 만들기
def even_num(num_list):
    if type(num_list) is list:
        result = [x for x in num_list if x % 2 == 0]
    else:
        result = []
        print('Your input is {}. Please input list'.format(type(num_list)))
    return result

# 함수 사용
a = 1
a = even_num(a)
a
```

## 8.2. 함수 입력과 출력

- 함수 입력과 출력은 매개변수와 리턴값을 통해 이루어집니다.

### 8.2.1. 여러 유형의 함수

- 함수가 항상 매개변수와 리턴값을 갖는 것은 아닙니다.

**1) 매개변수와 리턴값이 있는 함수**

```python
# 두 매개변수의 합을 반환하는 함수
def hap(x, y):
    r = x + y
    return(r)

# 함수 사용
a = hap(10, 20)
a
```

**2) 매개변수는 없고 리턴값만 있는 함수**

```python
# 매개변수 없이 값을 반환하는 함수
def hello():
    return '안녕하세요!'

# 함수 사용
a = hello()
a
```

**3) 매개변수는 있고 리턴값이 없는 함수**

```python
# 매개변수를 받아 값을 출력하는 함수
def hello(who):
    print(who + '님 안녕하세요!')

# 함수 사용
hello('홍길동')
```

```python
# 반환되는 값이 없음을 확인
def hello(who):
    print(who + '님 안녕하세요!')

# 함수 사용
a = hello('홍길동')
print(a)
```

**4) 매개변수와 리턴값이 없는 함수**

```python
# 매개변수와 반환되는 값이 없는 함수
def hello():
    print('여러분 안녕하세요!')

# 함수 사용
hello()
```

### 8.2.2. 여러 매개변수가 있는 함수

- 일반적으로 여러 개의 매개변수가 있는 경우 함수 안에서 반복문을 사용해 처리합니다.

```python
# 임의의 여러 매개변수가 있는 함수
def hello(*args):
    for who in args:
        print('{}님 안녕하세요!'.format(who))

# 함수 사용
hello('홍길동', '일지매', '강우동')
```

```python
# 임의의 여러 매개변수와 고정된 매개변수가 있는 함수
def hello(*args, title):
    for who in args:
        print('{0}{1} 안녕하세요!'.format(who, title))

# 함수 사용
hello('홍길동', '일지매', '강우동', title='선생님')
```

### 8.2.3. 리턴값이 여러개인 함수

- 리턴값이 여러개인 경우 튜플 형태로 반환됩니다.

**1) 리턴값이 두 개**

```python
# 리턴값이 두 개인 함수
def hap2(x, y):
    return x + y, x * y

# 함수 사용
r = hap2(10, 20)                 
r
```

**2) 리턴값이 세 개**

```python
# 리턴값이 세 개인 함수
def hap2(x, y):
    return x + y, x - y, x * y

# 함수 사용
r = hap2(10, 20)                 
r
```

**3) 여러 변수에 리턴값 대입**

- 이럴 때 튜플이 능력을 발휘합니다.

```python
# 리턴을 두 변수에 각각 받는 방법
def hap2(x, y):
    return x + y, x * y

# 함수 사용
s, m = hap2(10, 20)
print(s, m)
```

### 8.2.4. 매개변수 기본값을 갖는 함수

- 매개변수가 기본값을 가지면 함수를 호출할 때 매개변수를 전달하지 않아도 됩니다.
- 맨 뒷쪽 매개변수부터 기본값을 가질 수 있습니다.

```python
# 매개변수 기본값을 갖는 함수
def hello(name, loud=False):      # 디폴트 파라미터는 순서상 뒤쪽
    if loud:
        print('HELLO, {}!'.format(name.upper()))
    else:
        print('Hello, {}'.format(name))

# 함수 사용
hello('Bob')
hello('Fred', loud=True)
hello('James', loud=False)
```

```python
# 매개변수 기본값을 갖는 함수
def hello(who, title, end='!'):
    print('{0}{1} 안녕하세요{2}'.format(who, title, end))

# 함수 사용
hello('홍길동', '선생님', '?')
hello('홍길동', '선생님')
```

```python
# 중간의 매개변수는 기본값을 설정할 수 없음
def hello(who, end='!', title):
    print('{0}{1} 안녕하세요{2}'.format(who, title, end))

# 함수 사용
hello('홍길동', '선생님')
```

## 8.3. 함수와 변수

- 함수 안에서는 밖에서 선언한 변수를 변경할 수 없습니다.

```python
# 함수 안에서 밖의 변수를 변경할 수 없음
s1 = '홍길동'

def hello(s1):
    s1 = s1 + '님 안녕하세요!'

# 함수 사용
hello(s1)
s1
```

- 함수가 반환하는 값을 함수 밖의 변수가 받는 방법이 있습니다.

```python
# 반환되는 값을 받아 사용하기
s1 = '홍길동'

def hello(s1):
    return s1 + '님 안녕하세요!'

# 함수 사용
s1 = hello(s1)
s1
```

- global을 사용하면 함수 밖의 변수를 변경할 수 있습니다.

```python
# global 사용
s1 = '홍길동'

def hello():
    global s1
    s1 = s1 + '님 안녕하세요!'

# 함수 사용
hello()
s1
```
