- 배열의 크기(Shape)를 확인할 수 있습니다.
- Reshape 기능으로 배열 크기를 바꿀 수 있습니다.

# P1. [실습] 배열 만들기

```python
# 라이브러리 불러오기
import numpy as np
```

1) a 리스트를 사용해 **np_01** 배열을 생성하세요

```python
# 리스트 만들기
a = [1, 2, 3, 4, 5, 6]

# 배열 만들기
np_01 = np.array(a)

# 확인
print(np_01)
```

2) np_01의 크기를 확인하세요.

```python
# 배열 크기 확인
np_01.shape
```

3) np_01의 크기를 **(6, 1)**로 변환하여 **np_02**에 넣으세요.

```python
# 6 x 1 형태로 변경
np_02 = np_01.reshape(6,1)
print(np_02)
```

4) np_01의 크기를 **(1, 6)**로 변환하여 **np_03**에 넣으세요.

```python
# 1 x 6 형태로 변경
np_03 = np_01.reshape(1, 6)
print(np_03)
```

5) np_01의 크기를 **(-1, 2)**로 변환하여 **np_04**에 넣으세요.

```python
# -1 x 2 형태로 변경
np_04 = np_01.reshape(-1, 2)
print(np_04)
```

6) np_01, np_02, np_03, np_04의 **크기**와 **값**을 확인하세요.

```python
# np_01 크기 확인
print(np_01.shape)

# np_01 값 확인
print(np_01)
```

```python
# np_02 크기 확인
print(np_02.shape)

# np_02 값 확인
print(np_02)
```

```python
# np_03 크기 확인
print(np_03.shape)

# np_03 값 확인
print(np_03)
```

```python
# np_04 크기 확인
print(np_04.shape)

# np_04 값 확인
print(np_04)
```

7) 랜덤값으로 채운 **12 x 6** 형태의 2차원 배열을 생성하여 **np_sample** 변수로 선언하세요.

```python
# 12 x 6 형태의 2차원 배열 만들기
np_sample = np.random.randint(-24, 24, (12, 6))

# 크기 확인
print(np_sample.shape)

# 값 확인
print(np_sample)
```

8) np_sample 배열을 24 x ? 형태로 변경하세요.

```python
# 24 x ? 형태의 2차원 배열 만들기
np_sample2 = np_sample.reshape(24, -1)

# 크기 확인
print(np_sample2.shape)

# 값 확인
print(np_sample2)
```
