# P2. [실습] 데이터프레임 조회

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

1) 다음 파일을 읽어 **graduate**, **titanic** 데이터프레임을 만드세요.
- 'https://raw.githubusercontent.com/Jangrae/csv/master/Graduate_apply.csv'
- 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_simple.csv'

```python
# graduate 데이터프레임 만들기
g_path = 'https://raw.githubusercontent.com/Jangrae/csv/master/Graduate_apply.csv'
graduate = pd.read_csv(g_path)
graduate.head()
```

```python
# titanic 데이터프레임 만들기
t_path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_simple.csv'
titanic = pd.read_csv(t_path)
titanic.head()
```

### graduate 데이터프레임 사용

2) **unique()** 메소드를 사용해 rank 열에 어떤 값들이 들어 있는 지 확인하세요.

```python
# rank 값 확인
graduate['rank'].unique()
```

3) **mean()** 메소드를 사용해 gre 열의 평균을 구하세요.

```python
# gre 평균
graduate['gre'].mean()
```

4) gre 값이 평균 이상이면서 rank가 4인 사람을 조회하세요.

```python
# gre 값이 gre 평균 이상이고 rank가 4인 사람
graduate[(graduate['gre'] >= graduate['gre'].mean()) & (graduate['rank'] == 4)]
```

## titanic 데이터프레임 사용

5) Age가 10살 이하인 탑승객을 조회하세요.

```python
# Age가 10 이하인 탑승객 조회
titanic[titanic['Age'] <= 10]
```

6) 여자(Female)이면서, 탑승지역(Embarked)이 Southhampton 인 사람을 조회하세요.

```python
# 조건에 맞는 행 조회
titanic[(titanic['Sex'] == 'female') & (titanic['Embarked'] == 'Southhampton')]
```

7) 6번에 해당되는 사람의 평균 나이는 어떻게 되나요?

```python
# 조건에 맞는 행의 특정 열 평균 구하기
titanic[(titanic['Sex'] == 'female') & (titanic['Embarked'] == 'Southhampton')]['Age'].mean()
```

```python
titanic.loc[(titanic['Sex'] == 'female') & (titanic['Embarked'] == 'Southhampton'), ['Age']].mean()
```
