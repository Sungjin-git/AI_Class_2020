# P3. 데이터프레임 변경

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

1) 'titanic_simple.csv' 파일을 불러와 데이터프레임 **titanic**을 만드세요.

- 파일 경로: 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_simple.csv'

```python
# 파일 읽어오기
f_path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_simple.csv'
titanic = pd.read_csv(f_path)
titanic.head()
```

2) **drop()** 메소드를 사용해 Name 열을 삭제하세요.

```python
# Name 열 삭제
titanic.drop('Name', axis=1, inplace=True)
titanic.head()
```

3) **pd.get_dummies()** 함수로 Pclass 열의 Dummy Variable로 변환한 후  **pd.concat()** 함수로 기존 데이터 셋에 붙이세요.

```python
titanic['Pclass'].value_counts()
```

```python
# Dummy Variable 만들기
titanic_dummy = pd.get_dummies(titanic['Pclass'])
titanic_dummy.head()
```

```python
# 기존 데이터프레임에 붙이기
titanic_new = pd.concat([titanic, titanic_dummy], axis=1)
titanic_new.head()
```

4) **rename()** 메소드를 사용해 Dummy Variable로 추가한 열 이름을 Pclass_1, Pclass_2, Pclass_3 로 바꾸세요.

```python
# 열 이름 변경
titanic_new.rename(columns={1: 'Pclass_1',
                            2: 'Pclass_2',
                            3: 'Pclass_3'}, inplace=True)
titanic_new.head()
```

5) **value_counts()** 메소드를 사용해 Age 열에 NaN 값을 포함해서 어떤 값이 몇 개씩 있는지 확인하세요.

```python
# Age 열 값 개수 확인
titanic_new['Age'].value_counts(dropna=False)
```

6) **dropna()** 메소드로 Age 열이 NaN인 행은 삭제한 후 위 5)번 코드를 수행해 NaN값이 없어짐을 확인하세요.

```python
# Age 열 값이 NaN인 행 삭제
titanic_new.dropna(subset=['Age'], inplace=True)
titanic_new['Age'].value_counts(dropna=False)
```

7) **info()** 메소드를 사용해 아직도 NaN 값이 있는 열이 있는지 확인하세요.

```python
# 행 정보 확인
titanic_new.info()
```

8) **fillna()** 메소드를 사용해 Embarked 열의 NaN 값은 Southhampton로 채우세요.

```python
# NaN 값 채우기
titanic_new['Embarked'].fillna('Southhampton', inplace=True)
titanic_new.head()
```

9) **value_counts()** 메소드를 사용해서 Embarked 열에 NaN 값이 있는지 확인하세요.

```python
# 확인
titanic_new['Embarked'].value_counts(dropna=False)
```
