## 17. 롤링 & 시프트

- 우선 2015, 2016년 월 판매 수량 합을 집계해서 **sales_new** 데이터프레임을 만듭니다.
- groupby()

- 결과를 확인합니다.

- 현재 qty와 하루 전 qty 값을 비교할 것이면 qty 값을 하나씩 아래에 표시하면 됩니다.
- **shift()** 메소드에 이동할 행 수를 지정하면 됩니다.
- shift한 결과를 **prev_qty**열로 추가합니다.

```python
sales_new['prev_qty'] = sales_new['qty'].shift(1)
```

- 현재 월을 포함해서 최근 3개월 판매 수량 평균을 **lag3_qty_1** 열로 추가합니다.

```python
sales_new['lag3_qty_1'] = sales_new['qty'].rolling(3, min_periods=1).mean()
```

- 현재 월을 포함하지 않은 직전 3개월 판매 수량 평균을 **lag3_qty_2** 열로 추가합니다.

```python
sales_new['lag3_qty_2'] = sales_new['qty'].rolling(3, min_periods=1).mean().shift(1)
```

- 결과를 확인합니다.

```python
sales_new
```

## 18. 결측치 처리 #2

- 집계 과정에서 **lag3_qty_2** 열 맨 앞에 여전히 NaN 값이 있습니다.
- **fillna()** 메소드에 **method='bfill'**를 지정하여 바로 뒤의 값으로 채웁니다.

```python
sales_new.fillna(method='bfill', inplace=True)
```

- 결과를 확인합니다.

```python
sales_new
```

## 19. 반올림

- lag3_qty_1, lag3_qty_2 열 값 소수점이 다소 거슬립니다.
- 판매 수량이니 소수점을 없애기 위해 **round()** 메소드로 **반올림** 합니다.

```python
sales_new['lag3_qty_1'] = sales_new['lag3_qty_1'].round(0)
sales_new['lag3_qty_2'] = sales_new['lag3_qty_2'].round(0)
```

- 결과를 확인합니다.
