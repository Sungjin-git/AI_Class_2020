### 리스트

```
# 리스트 선언
#    0  1  2  3  4  5  6  7  8
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
a
```

```
# 인덱싱
a[-1]    # 9
```

```
# 슬라이싱
a[2:6]    # [3, 4, 5, 6]
```

```
# 맨 뒤에 10 추가
a.append(10)
a
```

```
# 맨 뒤에 11 추가
a = a + [11]
a
```

```
# 참고 : extend, insert
```

```
# range 함수
a = list(range(1, 11))

# a 요소 중에서 짝수만 출력 (for 문 사용)
for num in a:
    if num % 2 == 0:
        print(num)
```

```
# a 요소의 합
sum(a)
```

```
max(a)
```

```
# a의 평균
sum(a) / len(a)
```

### 딕셔너리

```
week = {'영어': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
        '한글': ['월', '화', '수', '목', '금', '토', '일']}
week
```

```
week['영어']
```

```
week['한글'][-1]
```

```
# 월(Mon), 화(Tue) ...
for day in range(len(week['한글'])):
    print('{}({})'.format(week['한글'][day], week['영어'][day]))
```
