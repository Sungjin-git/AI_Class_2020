```
def same(x, y):
    return list(set(x) & set(y))
```

```
a = [1, 2, 3, 4, 5, 6]
b = [3, 4, 5, 6, 7, 8]
c = same(a, b)
c
```

```
a, b = b, a
print(a, b)
```
