```
a = list(range(10))
a
```

```
# 짝수만 추출해서 b에 넣기
b = []
for x in a:
    if x % 2 == 0:
        b.append(x)
b
```

```
# 확장구문 --> List Comprehension 사용
b = [x for x in a if x % 2 == 0]
b
```

```
# 라이브러리 불러오기
import numpy as np
```

```
# 1000 ~ 2000 사이의 임의의 정수를 갖는 6 x 5 형태의 배열 만들기
# rand 0 ~ 1 균일 분포
# randn 가우시안 분포
# randint 정수 범위에서 균일 분포
a = np.random.randint(1000, 2000, (6, 5))
print(a)
```

```
# 배열 a를 100으로 나누기
a = a / 100
print(a)
```

```
# 요소의 데이터 형식
a.dtype
```

```
# int로 변경
a = a.astype(int)
print(a)
a.dtype
```

```
# 6 x 5 --> 3 x 10
a = a.reshape(3, -1)
print(a)
```

```
# shape 확인
a.shape
```

```
# Pandas 라이브러리 불러오기
import pandas as pd
```

```
# 시리즈 만들기
my_data = np.random.rand(10) * 100
my_date = pd.date_range('2020-05-01', periods=10)
s = pd.Series(my_data, index=my_date)
s
```

```
# 라인 그래프 그리기
s.plot()
```

```
# matplotlib 라이브러리 불러오기
import matplotlib.pyplot as plt
```

```
s.plot(kind='bar')
plt.show()
```
