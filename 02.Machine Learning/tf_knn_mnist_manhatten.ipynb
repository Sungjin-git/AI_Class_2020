```
# KNN 예제 - MNIST
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
```

```
n_training_img = 5000
n_testing_img = 200

pixel_train, onehot_train = mnist.train.next_batch(n_training_img) #(5000, 784), (5000, 10)
pixel_test, onehot_test = mnist.train.next_batch(n_testing_img)    #(200, 784), (200, 10)
TRAIN = tf.placeholder('float', [None, 784])
TEST = tf.placeholder('float', [784])
```

```
d = tf.reduce_sum(tf.abs(tf.add(TRAIN, tf.negative(TEST))), reduction_indices=1) # 거리 계산, (5000, 784)
val, idx = tf.nn.top_k(-d, k=5, sorted=False) # 제일 가까운 5개를 뽑기 위해 d에 마이너스를 붙인다
```

```
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    acc = 0.0

    for i in range(n_testing_img):
        knn_idx = sess.run(idx, feed_dict={TRAIN: pixel_train, TEST: pixel_test[i,:]})
        look_up = np.zeros(10)

        # print(knn_idx)
        # print(onehot_train[knn_idx])
        # print(np.argmax(onehot_train[knn_idx], axis=1))

        for j in np.argmax(onehot_train[knn_idx], axis=1):
            look_up[j] += 1
        # print(look_up)
        pred = np.argmax(look_up)

        print('Test: ', i, 'Prediction: ', pred, 'Real: ', np.argmax(onehot_test[i]))

        if pred == np.argmax(onehot_test[i]):
            acc += 1./n_testing_img
    
    print('Accuracy: ', acc)
```
