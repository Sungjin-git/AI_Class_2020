# Machine Learning for Modeling (과정평가)

```
!pip install tensorflow==1.5.0
# 문제 1] MNIST를 가져오는 부분 코딩하기
# 문제 2] 중간에 비선형함수를 추가하는데, 행렬이 784-by-500이 되도록(신경망에서 유닛500개)
#         1번 추가하기
# 문제 3] softmax의 cost함수를 있는 것을 쓰지말고, 강의노트 cross entropy공식보도 직접 구현
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)

import tensorflow as tf
import time
X = tf.placeholder(tf.float32, shape=[None,784])
Y = tf.placeholder(tf.float32, shape=[None,10])

W1 = tf.Variable(tf.random_normal([784, 500]))
b1 = tf.Variable(tf.random_normal([500]))
W2 = tf.Variable(tf.random_normal([500, 10]))
b2 = tf.Variable(tf.random_normal([10]))

batch_size = 128
num_steps = 5000

# model, cost, train
LC = tf.sigmoid(tf.matmul(X, W1)+b1)
model_LC = tf.matmul(LC, W2)+b2
model = tf.nn.softmax(model_LC)
#cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=model_LC, labels=Y))
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(model), axis=1))
train = tf.train.AdamOptimizer(0.01).minimize(cost)

acc = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(model, 1), tf.argmax(Y,1)), tf.float32))
# Session
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  # Training
  t1 = time.time()
  for step in range(1,num_steps+1):
    train_img, train_lbl = mnist.train.next_batch(batch_size)
    c, _ = sess.run([cost, train], feed_dict={X: train_img, Y: train_lbl})
    if step % 500 == 0:
      print(step, c)
  t2 = time.time()
  print("Training Time: ", t2-t1)
  # Testing
  print(sess.run(acc, feed_dict={X: mnist.test.images, Y: mnist.test.labels}))
```
