# Binary Classification

```
import tensorflow as tf
# X, Y 입력
x_data = [[1,2], [2,3], [3,1], [4,3], [5,3], [6,2]]
y_data = [[0], [0], [0], [1], [1], [1]]
X = tf.placeholder('float', [None, 2])
Y = tf.placeholder('float', [None, 1])
# W, b 변수
W = tf.Variable(tf.random_normal([2, 1]))
b = tf.Variable(tf.random_normal([1]))
# model, cost, train
model = tf.sigmoid(tf.matmul(X, W) + b)
cost = tf.reduce_mean((-1)*Y*tf.log(model) + (-1)*(1.0-Y)*tf.log(1.0-model))
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
# predict, accuracy
prediction = tf.cast(model > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, Y), dtype=tf.float32))
# Session 생성
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    # Training
    for step in range(10001):
        c, _ = sess.run([cost, train], feed_dict={X: x_data, Y: y_data})
        if step % 1000 == 0:
            print(step, c)
    # Test
    h, p, a = sess.run([model, prediction, accuracy], feed_dict={X: x_data, Y: y_data})
    print('Model:\n', h, '\nPrediction:\n', p, '\nAccuarcy:\n', a)

# 파라미터 : W, b 
# 하이퍼 파라미터 : learning rate, traing 횟수 등
# 하이퍼 파라미터 튜닝 필요
```
