# Linear Regression

```
import tensorflow as tf

# X, Y 입력 정의
x_data = [[1,1], [2,2], [3,3]]
y_data = [[10], [20], [30]]
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])
# placeholder는 다양한 입력 방법 가능, 
# ex) tf.placeholder('float', [None, 2])
# ex) tf.placeholder('float', (None, 2))

# W, b 변수 정의
W = tf.Variable(tf.random_normal([2,1]))
b = tf.Variable(tf.random_normal([1]))

# model, cost, train 정의
model = tf.matmul(X, W) + b
cost = tf.reduce_mean(tf.square(model - Y))    # MSE: Mean Squre Error
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)

# Session 생성
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    # Training
    for step in range(30001):
        c, _ = sess.run([cost, train], feed_dict={X: x_data, Y: y_data})
        if step % 1000 == 0:
            print(step, c)
    # Test
    print(sess.run(model, feed_dict={X: [[4,4]]}))
```
