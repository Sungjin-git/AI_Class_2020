# Softmax Classification

```
# minist 데이터셋 불러오는 코드는 꼭 기억하자!!!
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('mnist_data', one_hot=True)

import tensorflow as tf
import time
```

```
batch_size = 128
num_steps = 10000

X = tf.placeholder(tf.float32, shape=[None, 784])
Y = tf.placeholder(tf.float32, shape=[None, 10])

W = tf.Variable(tf.random_normal([784, 10]))
b = tf.Variable(tf.random_normal([10]))

model_LC = tf.matmul(X, W) + b
model = tf.nn.softmax(model_LC)
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=model_LC, labels=Y))
train = tf.train.AdamOptimizer(0.01).minimize(cost)

acc = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(model, 1), tf.argmax(Y, 1)), dtype=tf.float32))
```

```
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    # Training
    t1 = time.time()
    for step in range(num_steps):
        train_images, train_labels = mnist.train.next_batch(batch_size)  
        sess.run([cost, train], feed_dict={X: train_images, Y: train_labels})
    t2 = time.time()
    print('Training time: ', t2 - t1)

    # Teset
    # test_images, test_labels = mnist.test.next_batch(batch_size)  
    # print(sess.run(acc, feed_dict={X: test_images, Y: test_labels}))
    print(sess.run(acc, feed_dict={X: mnist.test.images, Y: mnist.test.labels}))
```
